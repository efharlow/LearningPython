```python
import numpy as np
import matplotlib.pyplot as plot
%matplotlib inline

def function(x):
    return x**2 + 4*x + 6
def derivative(x):
    return 2*x + 4
def descent(startX, dampening):
    points_tried = [(startX, function(startX))]
    def run(lastX, count):
        count -= 1
        if np.absolute(derivative(lastX)) < .1 or count < 0:
            return
        nextX = lastX - dampening * derivative(lastX)
        points_tried.append((nextX, function(nextX)))
        run(nextX, count)
    run(startX, 300)
    return points_tried
def plot_descent(points_tried):
    plot.figure()
    plot.plot([point[0] for point in points_tried], [point[1] for point in points_tried], "o")
    plot.show()

plot_descent(descent(20, .1))
plot_descent(descent(20, .5))
plot_descent(descent(20, .75))
plot_descent(descent(20, .98))
plot_descent(descent(20, 1))
plot_descent(descent(20, 1.01))
#plot_descent(descent(20, 1.98))
#plot_descent(descent(20, 2))
```
